```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import math

# traj_dict = pickle.load(open('dynamics_test.pkl', 'rb'))
traj_dict = pickle.load(open('dynamics_test_torque_fix.pkl', 'rb'))

nam_dict = pickle.load(open('./statics/nam_c_track.pkl', 'rb'))
```

```python
mx, Mx = min(nam_dict['x']), max(nam_dict['x']);dx = Mx-mx
my, My = min(nam_dict['y']), max(nam_dict['y']);dy = My-my

row = 8
col = row * (dy / dx)
```

```python
fig, ax = plt.subplots(4, 1, figsize=(8,8))
ax[0].plot(traj_dict['Ffx']);ax[0].set_title("Ffx")
ax[1].plot(traj_dict['Ffy']);ax[1].set_title('Ffy')
ax[2].plot(traj_dict['Frx']);ax[2].set_title('Frx')
ax[3].plot(traj_dict['Fry']);ax[3].set_title('Fry')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(5, 1, figsize=(8,8))
ax[0].plot(traj_dict['car_phi']);ax[0].set_title("phi")
ax[1].plot(traj_dict['car_vx']);ax[1].set_title('vx')
ax[2].plot(traj_dict['car_vy']);ax[2].set_title('vy')
ax[3].plot(traj_dict['torque']);ax[3].set_title('torque')
ax[4].plot(traj_dict['steer']);ax[4].set_title('steer')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(row, col))
for i, (tx, ty) in enumerate(zip(nam_dict['x'], nam_dict['y'])):
    plt.scatter(tx, ty, s=1, c='k')
for i, (cx, cy) in enumerate(zip(traj_dict['car_x'], traj_dict['car_y'])):
    # if i == 100:
    #     break
    plt.scatter(cx, cy, s=3, c='r')
fig.tight_layout()
```
